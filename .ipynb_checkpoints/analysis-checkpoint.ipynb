# Covid-19 India Cases Analysis

This notebook studies the time-series data of Covid-19 India data set with the following heads: date, time, state/union territory, confirmed Indian national, confirmed Foreign national, cured, deaths and confirmed(sum of confirmed Indian national and confirmed foreign national cases recorded).

The data is available as a `csv` file. Filename: `covid-19-india-data.csv`.

The study is done on the data that is available from January 2020 to November 2021.

Currently the notebook consists the study of total confirmed Indian national cases, total confirmed Foreign national cases, total number of people cured, total deaths, total confirmed cases, state-wise confirmed cases and its bar graph representation and line graph of increase in confirmed cases over the whole duration present in the given data of any one state (in this notebook Bihar is choosen arbitrarily).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid-19-india-data.csv")
```

```python
display(df)
```

```python
print(df.dtypes)
```

**Note:** Here the dtype of `ConfirmedIndianNational` and `ConfirmedForeignNational` is an object type. To obtain the sum of these two fields we choose to make a numpy array of the two fields and sum the values using a for loop.

In certain records the value is set to a dash (-) so we have to exclude them from the loop.

```python
confirmed_indian_national_arr = df["ConfirmedIndianNational"].to_numpy()
print(confirmed_indian_national_arr)
```

```python
total_confirmed_indian_national = 0
for x in confirmed_indian_national_arr:
    if x != '-':
        total_confirmed_indian_national += int(x)
print("Total Confirmed Indian National:", total_confirmed_indian_national)
```

```python
confirmed_foreign_national_arr = df["ConfirmedForeignNational"].to_numpy()
print(confirmed_foreign_national_arr)
```

```python
total_confirmed_foreign_national = 0;
for x in confirmed_foreign_national_arr:
    if x != '-':
        total_confirmed_foreign_national += int(x)
print("Total Confirmed Foreign National:", total_confirmed_foreign_national)
```

```python
cured_arr = df["Cured"].to_numpy()
total_cured = np.sum(cured_arr)
print("Total Cured:", total_cured)
```

```python
deaths_arr = df["Deaths"].to_numpy()
total_deaths = np.sum(deaths_arr)
print("Total Deaths:", total_deaths)
```

```python
confirmed = df["Confirmed"].to_numpy()
total_confirmed = np.sum(confirmed)
print("Total Confirmed:", total_confirmed)
```

## Grouping state-wise confirmed cases

```python
state_wise_confirmed_cases = np.array(df.groupby("State/UnionTerritory")["Confirmed"].sum())
unique_states = df["State/UnionTerritory"].unique()
for i in range(len(unique_states)):
    print(f"{unique_states[i]}: {state_wise_confirmed_cases[i]}")
```

## Creating a dictionary of states and confirmed cases

```python
state_wise_confirmed_cases_dict = {"States": unique_states, "ConfirmedCases": state_wise_confirmed_cases}
print(state_wise_confirmed_cases_dict)
```

## Creating a state-wise dataframe from the above created dictionary

```python
state_wise_df = pd.DataFrame(state_wise_confirmed_cases_dict)
display(state_wise_df)
```

## Dropping the 27th index which are Unassigned cases
The unassigned cases are removed from the dataframe and the original dataframe is modified so that the bar graph can be created without this record.

```python
state_wise_df.drop(27, inplace=True)
```

## Creating bar-graph of the state-wise cases
The dpi is increased to improve the resolution of xlabel and a randomly generated 45 colors array is used to represent each state or union territory.

The xlabel is rotated 90 degrees to make the state/union territory names readable.

```python
fig = plt.figure(dpi=200)
colors = ['#cde15f', '#a98dd1', '#e7f070', '#5ba29f', '#ff2c6d', '#669e4f', '#f9d4f4', '#d4b4e8', '#a77260', '#d6b1c6', '#44c9a9', '#ef811f', '#c6b8e6', '#ff6d05', '#3e77b2', '#2e9c6d', '#d3aa71', '#8cb6d1', '#d7e280', '#d890b8', '#69be9d', '#f162bd', '#b5c5b5', '#f0a5e5', '#b8c468', '#9ac5e5', '#b56d87', '#8fc472', '#fd7d7d', '#ff9b87', '#b9bb75', '#81b3c3', '#ca8ec6', '#d6e7e0', '#ed81a8', '#9e8996', '#eca6c8', '#e8c2f2', '#f1c3d3', '#be97c6', '#f4d4d4', '#f6e89e', '#81b6d7', '#e1d8f6', '#9c95a6']

plt.bar(state_wise_df["States"], state_wise_df["ConfirmedCases"], width=0.75, color=colors)

plt.xticks(rotation=90, fontsize=5)
plt.title("State-wise Covid-19 Cases")

plt.xlabel("States")
plt.ylabel("Confirmed Cases")

plt.show()
```

## Creating bihar time-series dataframe

```python
bihar_time_series_df = df[df["State/UnionTerritory"] == "Bihar"]
bihar_time_series_df
```

## The line graph is plotted between date and cases confirmed
The xlabel is set to hidden because the dates were overlapping on each other. The data set consists of 18,078 records.

```python
fig = plt.figure(dpi=200)
plt.plot(bihar_time_series_df["Date"], bihar_time_series_df["Confirmed"])

plt.xlabel("Jan 2020 to Nov 2021")
plt.ylabel("Confirmed Cases")

plt.xticks([])

plt.show()
```
